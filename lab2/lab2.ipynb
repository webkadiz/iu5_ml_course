```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

train_df = pd.read_csv('Gender_Inequality_Index.csv')

train_df.head(10)
```

```python
train_df.describe(include = "all")
```

```python
train_df.info()
```

```python
train_df.isnull().sum()
```

```python
train_df.drop(train_df[train_df['GII'].isnull()].index, inplace=True)
train_df.reset_index(inplace=True, drop=True)
```

```python
train_df.tail()
```

```python
onehotencoder = OneHotEncoder()

encoder_df = pd.DataFrame(onehotencoder.fit_transform(train_df[['Human_development']]).toarray())

encoder_df.columns = ['hd_high', 'hd_low', 'hd_medium', 'hd_veryhigh']

train_df = train_df.join(encoder_df)
```

```python
train_df.drop('Human_development', axis=1, inplace=True)
```

```python
train_df.drop(['Country'], axis=1, inplace=True)
```

```python
train_df.info()
```

```python
norm = MinMaxScaler().fit_transform(train_df)

train_df[train_df.columns] = norm
train_df.describe()
```
